Genellikle, ikili sınıflandırma görevlerinde normal durum olan bir sınıf ve anormal durum olan başka bir sınıf bulunur.

```python
# example of binary classification task
from numpy import where
from collections import Counter
from sklearn.datasets import make_blobs
from matplotlib import pyplot
# define dataset
X, y = make_blobs(n_samples=1000, centers=2, random_state=1)
# summarize dataset shape
print(X.shape, y.shape)
# summarize observations by class label
counter = Counter(y)
print(counter)
# summarize first few examples
for i in range(10):
 print(X[i], y[i])
# plot the dataset and color the by class label
for label, _ in counter.items():
 row_ix = where(y == label)[0]
 pyplot.scatter(X[row_ix, 0], X[row_ix, 1], label=str(label))
pyplot.legend()
pyplot.show()
```

**Tek Katmanlı Algılayıcı(Single Layer Perceptron)**

En eski ve ilk tanıtılan sinir ağlarından biridir. Frank Rosenblatt tarafından 1958 yılında önerilmiştir. Perceptron yapay sinir ağı olarak da bilinir. Perceptron temel olarak ikili girişi ve ikili çıkışı olan AND, OR ve NOR gibi mantıksal kapıları hesaplamak için kullanılır.

Algılayıcının ana işlevleri şunlardır:

Giriş katmanından girdi alır.
Bunları ağırlıklandırıp toplar.
Çıktıyı üretmek için toplamı doğrusal olmayan fonksiyona aktarır.

Perceptron temel olarak verileri iki parçaya sınıflandırmak için kullanılır. Bu nedenle Doğrusal İkili Sınıflandırıcı olarak da bilinir.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
iris = load_iris()
X, y = iris.data[:, :2], (iris.target != 0) * 1  # Take only the first two features and convert labels to binary

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardize features by removing the mean and scaling to unit variance
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create and train a Single Layer Perceptron (SLP)
slp = Perceptron(max_iter=1000, random_state=42)
slp.fit(X_train_scaled, y_train)

# Plot decision boundary
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))
Z = slp.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.4)
plt.scatter(X[:, 0], X[:, 1], c=y, s=20, edgecolor='k')
plt.title('Single Layer Perceptron Decision Boundary')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

**Çok Katmanlı Algılayıcı(Multi-layer Perceptron)**

Çok katmanlı algılama MLP olarak da bilinir. Herhangi bir girdi boyutunu istenen boyuta dönüştüren tamamen bağlı yoğun katmanlardır. Çok katmanlı bir algı, birden fazla katmana sahip bir sinir ağıdır. Bir sinir ağı oluşturmak için nöronları bir araya getiririz, böylece bazı nöronların çıktıları diğer nöronların girdileri olur.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = load_iris()
X, y = iris.data, iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardize features by removing the mean and scaling to unit variance
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create and train a Multi-Layer Perceptron (MLP) classifier
mlp = MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000, random_state=42)
mlp.fit(X_train_scaled, y_train)

# Predict on the test set
y_pred = mlp.predict(X_test_scaled)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Plot the decision boundary (2D visualization only)
# Note: This visualization is simplified and may not accurately represent the decision boundary of the MLP
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolor='k')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('MLP Decision Boundary (2D Visualization)')
plt.show()
```

1. **Basitlik ve Karmaşıklık**

SLP: Tek bir nöron katmanına sahip basit doğrusal sınıflandırıcı. Doğrusal karar sınırlarını öğrenmekle sınırlıdır. MLP: Birden fazla katmana (bir veya daha fazla gizli katman dahil) ve doğrusal olmayan aktivasyon fonksiyonlarına sahip daha karmaşık sinir ağı. Doğrusal olmayan karar sınırlarını öğrenebilir.
2. **Temsil Gücü**

SLP: Doğrusal yapısı nedeniyle sınırlı temsil gücü. Sadece girdi özellikleri ve çıktı arasındaki doğrusal ilişkileri modelleyebilir.
MLP: Doğrusal olmayan aktivasyon fonksiyonlarının ve çoklu katmanların varlığı nedeniyle daha yüksek temsil gücü. Verilerdeki karmaşık doğrusal olmayan ilişkileri modelleyebilir.
3. **Esneklik**

SLP: MLP'ye kıyasla daha az esnektir. Doğrusal karar sınırları ve doğrusal ilişkilerle sınırlıdır.
MLP: Daha esnek ve çok yönlüdür. Doğrusal olmayan karar sınırlarını öğrenebilir ve verilerdeki karmaşık örüntüleri yakalayabilir.
4. **Eğitim Karmaşıklığı**

SLP: Eğitim MLP'ye kıyasla daha basittir. Ayarlanması gereken gizli katmanlar veya doğrusal olmayan aktivasyon fonksiyonları yoktur.
MLP: Birden fazla katman, doğrusal olmayan aktivasyon fonksiyonları ve potansiyel olarak daha büyük parametre uzayının varlığı nedeniyle eğitim daha karmaşık olabilir.
5. **Performans**

SLP: Doğrusal olarak ayrılabilen basit veri kümelerinde veya doğrusal karar sınırlarının yeterli olduğu görevlerde iyi performans gösterir.
MLP: Doğrusal olmayan ilişkiler ve doğrusal olarak ayrılamayan veriler içeren karmaşık sınıflandırma görevlerinde genellikle daha iyi performans gösterir.
6. **Aşırı Uyum(Overfitting)**

SLP: Daha basit mimarisi nedeniyle MLP'ye kıyasla aşırı uyuma daha az eğilimlidir.
MLP: Özellikle küçük veri kümelerinde veya model mimarisi aşırı karmaşık olduğunda aşırı uyuma daha yatkındır. Düzenli hale getirme teknikleri genellikle MLP'lerde aşırı uyumu azaltmak için kullanılır.
7. **Yorumlanabilirlik**

SLP: Daha basit mimarisi ve doğrusal karar sınırları nedeniyle MLP'ye kıyasla daha yorumlanabilir.
MLP: Karmaşık mimarisi ve doğrusal olmayan aktivasyon fonksiyonları nedeniyle daha az yorumlanabilir. Öğrenilen temsillerin anlaşılması daha zor olabilir.
8. **Ölçeklenebilirlik**

SLP: Özellikle karmaşık görevler veya büyük veri kümeleri için sınırlı ölçeklenebilirlik.
MLP: Daha ölçeklenebilir ve mimariyi (örneğin, daha fazla katman, nöron ekleyerek) ve düzenleme tekniklerini ayarlayarak daha büyük ve daha karmaşık veri kümelerini işlemek için uyarlanabilir.

Bu karşılaştırmalar Tek Katmanlı Algılayıcı (SLP) ve Çok Katmanlı Algılayıcı (MLP) arasındaki basitlik, karmaşıklık, esneklik, performans ve diğer faktörler açısından farklılıkları vurgulamaktadır. İkisi arasındaki seçim, sınıflandırma görevinin doğasına, verilerin karmaşıklığına ve basitlik ile performans arasındaki değiş tokuşlara bağlıdır.

**Bir Sınıflandırma modelinin değerlendirilmesi**

Modelimiz tamamlandıktan sonra, performansını değerlendirmek gerekir; bu ya bir Sınıflandırma ya da Regresyon modelidir. Dolayısıyla, bir Sınıflandırma modelini değerlendirmek için aşağıdaki yollara sahibiz:

1. Log Kaybı veya Çapraz Entropi Kaybı(Log Loss or Cross-Entropy Loss)

Çıktısı 0 ile 1 arasında bir olasılık değeri olan bir sınıflandırıcının performansını değerlendirmek için kullanılır.
İyi bir ikili Sınıflandırma modeli için log kaybının değeri 0'a yakın olmalıdır.
Tahmin edilen değer gerçek değerden saparsa log kaybının değeri artar.
Düşük log kaybı, modelin daha yüksek doğruluğunu temsil eder.
İkili sınıflandırma için çapraz entropi şu şekilde hesaplanabilir:
$$
{-(ylog(p)+(1-y)log(1-p))}
$$
Burada y= Gerçek çıktı, p= öngörülen çıktı.

2. Karışıklık matrisi(Confusion Matrix)

Karışıklık matrisi bize çıktı olarak bir matris/tablo sunar ve modelin performansını açıklar.
Hata matrisi olarak da bilinir.
Matris, doğru tahminlerin ve yanlış tahminlerin toplam sayısına sahip özetlenmiş bir formdaki tahmin sonuçlarından oluşur. Matris aşağıdaki tabloya benzer:

Gerçekleşen Pozitif, Gerçekleşen Negatif,
Tahmini Pozitif, Doğru Pozitif, Yanlış Pozitif,
Tahmini Negatif, Yanlış Negatif, Doğru Negatif,
Makine Öğreniminde Sınıflandırma Algoritması

```python
import numpy as np
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

#Create the NumPy array for actual and predicted labels
actual = np.array(
  ['Dog','Dog','Dog','Not Dog','Dog','Not Dog','Dog','Dog','Not Dog','Not Dog'])
predicted = np.array(
  ['Dog','Not Dog','Dog','Not Dog','Dog','Dog','Dog','Dog','Not Dog','Not Dog'])

#Compute the confusion matrix
cm = confusion_matrix(actual,predicted)

#Plot the confusion matrix with the help of the seaborn heatmap
sns.heatmap(cm,
            annot=True,
            fmt='g',
            xticklabels=['Dog','Not Dog'],
            yticklabels=['Dog','Not Dog'])
plt.ylabel('Prediction',fontsize=13)
plt.xlabel('Actual',fontsize=13)
plt.title('Confusion Matrix',fontsize=17)
plt.show()
```

3. AUC-ROC eğrisi

ROC eğrisi Alıcı İşletim Karakteristikleri Eğrisi, AUC ise Eğri Altındaki Alan anlamına gelmektedir.
Sınıflandırma modelinin farklı eşiklerdeki performansını gösteren bir grafiktir.
Çok sınıflı sınıflandırma modelinin performansını görselleştirmek için AUC-ROC Eğrisini kullanırız.
ROC eğrisi, TPR (Doğru Pozitif Oranı) Y ekseninde ve FPR (Yanlış Pozitif Oranı) X ekseninde olmak üzere TPR ve FPR ile çizilir.

```python
import numpy as np
from sklearn .metrics import roc_auc_score

y_true = [1, 1, 0, 0, 1, 0]
y_pred = [0.95, 0.90, 0.85, 0.81, 0.78, 0.70]
auc = np.round(roc_auc_score(y_true, y_pred), 3)
print("Auc for our sample data is {}".format(auc))

true_positive_rate = [0.4, 0.8,  0.8, 1.0, 1]
false_positive_rate = [0, 0,  0, 0.2, 0.8]

plt.plot(false_positive_rate, true_positive_rate, 'o-', label='ROC')
plt.plot([0, 1], [0, 1], '--', color='grey', label='Worst case')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```
