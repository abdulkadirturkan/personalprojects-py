**Tek Katmanl Alglayc(Single Layer Perceptron)**

En eski ve ilk tantlan sinir alarndan biridir. Frank Rosenblatt tarafndan 1958 ylnda nerilmitir. Perceptron yapay sinir a olarak da bilinir. Perceptron temel olarak ikili girii ve ikili k olan AND, OR ve NOR gibi mantksal kaplar hesaplamak iin kullanlr.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
iris = load_iris()
X, y = iris.data[:, :2], (iris.target != 0) * 1  # Take only the first two features and convert labels to binary

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardize features by removing the mean and scaling to unit variance
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create and train a Single Layer Perceptron (SLP)
slp = Perceptron(max_iter=1000, random_state=42)
slp.fit(X_train_scaled, y_train)

# Plot decision boundary
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))
Z = slp.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.4)
plt.scatter(X[:, 0], X[:, 1], c=y, s=20, edgecolor='k')
plt.title('Single Layer Perceptron Decision Boundary')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

**ok Katmanl Alglayc(Multi-layer Perceptron)**

ok katmanl alglama MLP olarak da bilinir. Herhangi bir girdi boyutunu istenen boyuta dntren tamamen bal youn katmanlardr. ok katmanl bir alg, birden fazla katmana sahip bir sinir adr. Bir sinir a oluturmak iin nronlar bir araya getiririz, bylece baz nronlarn ktlar dier nronlarn girdileri olur.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = load_iris()
X, y = iris.data, iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardize features by removing the mean and scaling to unit variance
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create and train a Multi-Layer Perceptron (MLP) classifier
mlp = MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000, random_state=42)
mlp.fit(X_train_scaled, y_train)

# Predict on the test set
y_pred = mlp.predict(X_test_scaled)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Plot the decision boundary (2D visualization only)
# Note: This visualization is simplified and may not accurately represent the decision boundary of the MLP
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolor='k')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('MLP Decision Boundary (2D Visualization)')
plt.show()
```

1. **Basitlik ve Karmaklk**

SLP: Tek bir nron katmanna sahip basit dorusal snflandrc. Dorusal karar snrlarn renmekle snrldr. MLP: Birden fazla katmana (bir veya daha fazla gizli katman dahil) ve dorusal olmayan aktivasyon fonksiyonlarna sahip daha karmak sinir a. Dorusal olmayan karar snrlarn renebilir.
2. **Temsil Gc**

SLP: Dorusal yaps nedeniyle snrl temsil gc. Sadece girdi zellikleri ve kt arasndaki dorusal ilikileri modelleyebilir.
MLP: Dorusal olmayan aktivasyon fonksiyonlarnn ve oklu katmanlarn varl nedeniyle daha yksek temsil gc. Verilerdeki karmak dorusal olmayan ilikileri modelleyebilir.
3. **Esneklik**

SLP: MLP'ye kyasla daha az esnektir. Dorusal karar snrlar ve dorusal ilikilerle snrldr.
MLP: Daha esnek ve ok ynldr. Dorusal olmayan karar snrlarn renebilir ve verilerdeki karmak rntleri yakalayabilir.
4. **Eitim Karmakl**

SLP: Eitim MLP'ye kyasla daha basittir. Ayarlanmas gereken gizli katmanlar veya dorusal olmayan aktivasyon fonksiyonlar yoktur.
MLP: Birden fazla katman, dorusal olmayan aktivasyon fonksiyonlar ve potansiyel olarak daha byk parametre uzaynn varl nedeniyle eitim daha karmak olabilir.
5. **Performans**

SLP: Dorusal olarak ayrlabilen basit veri kmelerinde veya dorusal karar snrlarnn yeterli olduu grevlerde iyi performans gsterir.
MLP: Dorusal olmayan ilikiler ve dorusal olarak ayrlamayan veriler ieren karmak snflandrma grevlerinde genellikle daha iyi performans gsterir.
6. **Ar Uyum(Overfitting)**

SLP: Daha basit mimarisi nedeniyle MLP'ye kyasla ar uyuma daha az eilimlidir.
MLP: zellikle kk veri kmelerinde veya model mimarisi ar karmak olduunda ar uyuma daha yatkndr. Dzenli hale getirme teknikleri genellikle MLP'lerde ar uyumu azaltmak iin kullanlr.
7. **Yorumlanabilirlik**

SLP: Daha basit mimarisi ve dorusal karar snrlar nedeniyle MLP'ye kyasla daha yorumlanabilir.
MLP: Karmak mimarisi ve dorusal olmayan aktivasyon fonksiyonlar nedeniyle daha az yorumlanabilir. renilen temsillerin anlalmas daha zor olabilir.
8. **leklenebilirlik**

SLP: zellikle karmak grevler veya byk veri kmeleri iin snrl leklenebilirlik.
MLP: Daha leklenebilir ve mimariyi (rnein, daha fazla katman, nron ekleyerek) ve dzenleme tekniklerini ayarlayarak daha byk ve daha karmak veri kmelerini ilemek iin uyarlanabilir.

```python
import numpy as np
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

#Create the NumPy array for actual and predicted labels
actual = np.array(
  ['Dog','Dog','Dog','Not Dog','Dog','Not Dog','Dog','Dog','Not Dog','Not Dog'])
predicted = np.array(
  ['Dog','Not Dog','Dog','Not Dog','Dog','Dog','Dog','Dog','Not Dog','Not Dog'])

#Compute the confusion matrix
cm = confusion_matrix(actual,predicted)

#Plot the confusion matrix with the help of the seaborn heatmap
sns.heatmap(cm,
            annot=True,
            fmt='g',
            xticklabels=['Dog','Not Dog'],
            yticklabels=['Dog','Not Dog'])
plt.ylabel('Prediction',fontsize=13)
plt.xlabel('Actual',fontsize=13)
plt.title('Confusion Matrix',fontsize=17)
plt.show()
```

3. AUC-ROC erisi

ROC erisi Alc letim Karakteristikleri Erisi, AUC ise Eri Altndaki Alan anlamna gelmektedir.
Snflandrma modelinin farkl eiklerdeki performansn gsteren bir grafiktir.
ok snfl snflandrma modelinin performansn grselletirmek iin AUC-ROC Erisini kullanrz.
ROC erisi, TPR (Doru Pozitif Oran) Y ekseninde ve FPR (Yanl Pozitif Oran) X ekseninde olmak zere TPR ve FPR ile izilir.

```python
import numpy as np
from sklearn .metrics import roc_auc_score

y_true = [1, 1, 0, 0, 1, 0]
y_pred = [0.95, 0.90, 0.85, 0.81, 0.78, 0.70]
auc = np.round(roc_auc_score(y_true, y_pred), 3)
print("Auc for our sample data is {}".format(auc))

true_positive_rate = [0.4, 0.8,  0.8, 1.0, 1]
false_positive_rate = [0, 0,  0, 0.2, 0.8]

plt.plot(false_positive_rate, true_positive_rate, 'o-', label='ROC')
plt.plot([0, 1], [0, 1], '--', color='grey', label='Worst case')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```
